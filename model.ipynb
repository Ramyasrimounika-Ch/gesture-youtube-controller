```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:/youtube_ai/final_dataset.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
X = df.drop('label', axis=1).values  # shape: (num_samples, 42)
y = df['label'].values
```

```python
y
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
y_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(42,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(9, activation='softmax')  
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test), batch_size=32)
```

```python
prediction_lastlayer=model.predict(X_test)
```

```python
prediction_lastlayer
```

```python
model.evaluate(X_test,y_test)
```

```python
y_classes = [
    "Left",  # 0
    "Right",       # 1
    "Volume down",            # 2
    "Volume up",           # 3
    "pause"            
]
```

```python
sample = X_test[500].reshape(1, -1)
prediction = model.predict(sample)
predicted_class = np.argmax(prediction)
print("predicted_class:",y_classes[predicted_class])
print("actual_class:",y_classes[y_test[500]])
```
